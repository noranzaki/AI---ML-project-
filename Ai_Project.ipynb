```python
import pandas as pd
import numpy as np                          
import matplotlib.pyplot as plt     

train=pd.read_csv('tmdb-movies (2).csv')
train.info()
```

```python
# Import label encoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in columns
train['cast']= label_encoder.fit_transform(train['cast'])
train['homepage']= label_encoder.fit_transform(train['homepage'])
train['director']= label_encoder.fit_transform(train['director'])
train['tagline']= label_encoder.fit_transform(train['tagline'])
train['keywords']= label_encoder.fit_transform(train['keywords'])
train['genres']= label_encoder.fit_transform(train['genres'])
train['production_companies']= label_encoder.fit_transform(train['production_companies'])

# Net Profit Column
train['net_profit']=train['revenue_adj']-train['budget_adj']
```

```python
train.corr()
```

```python
#Dropping useless columns
train = train.drop(columns = ['id'],axis=1)
train = train.drop(columns = ['imdb_id'],axis=1)
train = train.drop(columns = ['original_title'],axis=1)
train = train.drop(columns = ['homepage'],axis=1)
train = train.drop(columns = ['tagline'],axis=1)
train = train.drop(columns = ['keywords'],axis=1)
train = train.drop(columns = ['overview'],axis=1)
train = train.drop(columns = ['runtime'],axis=1)
train = train.drop(columns = ['release_year'],axis=1)
train = train.drop(columns = ['release_date'],axis=1)
train = train.drop(train[train.budget_adj <= 0].index,axis=0) 
train = train.drop(train[train.revenue_adj <= 0].index,axis=0) 
```

```python
#Checking for null values
train.isnull().sum()
```

```python
#Checking for Duplicates
train.duplicated().sum()
```

```python
train.drop_duplicates(inplace=True)
```

```python
train
```

```python
#Normalizing Values Function
def normalize():
    global train
    result = train.copy()
    for feature_name in train.columns:
        max_value = train[feature_name].max()
        min_value = train[feature_name].min()
        result[feature_name] = (train[feature_name] - min_value) / (max_value - min_value)
    train= result
```

```python
normalize()
```

```python
#Splitting Data to train and test
X= train.drop(['budget_adj','revenue_adj','net_profit'],axis=1)
Y= train['net_profit']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=3)

#Reshaping
Y_train = np.array(Y_train)
Y_train = Y_train.reshape(Y_train.shape[0], 1)

Y_test = np.array(Y_test)
Y_test = Y_test.reshape(Y_test.shape[0], 1)

X_train = np.vstack((np.ones((X_train.shape[0], )), X_train.T)).T
X_test = np.vstack((np.ones((X_test.shape[0], )), X_test.T)).T
```

```python
def model(X, Y, learning_rate, iteration):
  m = Y.size
  theta = np.zeros((X.shape[1], 1))
  cost_list = []
  for i in range(iteration):
    y_pred = np.dot(X, theta)
    cost = (1/(2*m))*np.sum(np.square(y_pred - Y))
    d_theta = (1/m)*np.dot(X.T, y_pred - Y)
    theta = theta - learning_rate*d_theta
    cost_list.append(cost)
    # to print the cost for 10 times
    if(i%(iteration/10) == 0):
      print("Cost is :", cost)
  return theta, cost_list
```

```python
iteration = 10000
learning_rate = 0.00005
theta, cost_list = model(X_train, Y_train, learning_rate = learning_rate, iteration =iteration)
```

```python
y_pred = np.dot(X_test, theta)
error = (1/X_test.shape[0])*np.sum(np.abs(y_pred - Y_test))
```

```python
print("Test error is :", error*100, "%")
print("Test Accuracy is :", (1- error)*100, "%")
```
